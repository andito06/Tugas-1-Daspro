```python
berat_badan = int(input("Masukkan berat badan anda: "))

if berat_badan <= 30:
    print("Ukuran Baju S")
elif berat_badan <= 40:
    print("Ukuran Baju M")
elif berat_badan <= 50:
    print("Ukuran Baju L")
elif berat_badan <= 65:
    print("Ukuran Baju XL")
elif berat_badan < 0 or berat_badan > 1000:
    print("Berat Badan Tidak Valid")
else:
    print("Tidak ada ukuran untuk berat tersebut")
```
